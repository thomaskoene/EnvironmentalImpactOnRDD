```python
# moving images to data folder

countries = ['China_Drone', 'China_MotorBike', 'Czech', 'India', 'Japan', 'Norway', 'United_States']

destination = "/gpfs/home5/tkoene/all_countries/data/images/train/"
for country in countries:
    # copy all images from the folders to one
    source = "/gpfs/home5/tkoene/RDD2022/" + country + "/data/images/train/"
    contents = os.listdir(source)
    for item in contents:
        if str(item) != ".ipynb_checkpoints":
            shutil.copyfile(source + str(item), destination + str(item))
```

```python
# splitting into train and val set (for weather circumstances)

# select 20% images random for every weather category
# move these from trainingsset to different folder

df = pd.DataFrame()

weather_categories = ['cloudy', 'rain', 'shine', 'sunrise']
for weather_category in weather_categories:
    images_path = "/gpfs/home5/tkoene/weather_classification/weather_models/" + weather_category + "/data/images/train/"
    labels_path = "/gpfs/home5/tkoene/weather_classification/weather_models/" + weather_category + "/data/labels/train/"
    

    images_files = [f for f in os.listdir(images_path) if isfile(join(images_path, f))]
    
    # apply filters
    # images_files = [i for i in images_files if not 'Norway' in i]
    # images_files = [i for i in images_files if not 'Czech' in i]
    
    sample = random.sample(images_files, int(len(image_files)*0.2))
    df[weather_category] = sample
        
    for file in sample:
        images_source = "/gpfs/home5/tkoene/all_countries/data/images/train/" + file
        labels_source = ("/gpfs/home5/tkoene/all_countries/data/labels/train/" + file).replace("jpg", "txt")
        
        images_destination = "/gpfs/home5/tkoene/all_countries/data/images/val/" + file
        labels_destination = ("/gpfs/home5/tkoene/all_countries/data/labels/val/" + file).replace("jpg", "txt")
        
        # print(images_source, labels_source)
        # print(images_destination, labels_destination)
        
        shutil.move(images_source, images_destination)
        shutil.move(labels_source, labels_destination)

df.to_csv('val_set_weather_category.csv')
```
