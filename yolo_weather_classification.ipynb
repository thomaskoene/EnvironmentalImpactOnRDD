```python
# Load a model
model = YOLO("yolov8n-cls.pt")  # load a pretained model

# Use the model
results = model.train(data='/gpfs/home5/tkoene/weather_classification/dataset', epochs=10)  # train the model
```

```python
model = YOLO("/gpfs/home5/tkoene/weather_classification/runs/classify/train5/weights/last.pt")
```

```python
countries = ['China_Drone', 'China_MotorBike', 'Czech', 'India', 'Japan', 'Norway', 'United_States']
country = countries[2]

labels = pd.read_csv("/gpfs/home5/tkoene/RDD2022/" + country + "/labels_train.csv")

labels = pd.DataFrame(labels['filename'].unique()).rename(columns={0: 'filename'})

def weather_pred(filename):
    results = model("/gpfs/home5/tkoene/RDD2022/" + country + "/data/images/train/" + str(filename))
    names_dict = results[0].names
    probs = results[0].probs.top1
    
    return names_dict[probs]

labels['weather'] = labels['filename'].map(weather_pred)

labels.to_csv('weather_labels_' + country + '.csv')
```
